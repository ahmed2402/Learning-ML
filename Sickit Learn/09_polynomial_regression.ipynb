# Polynomial Regression

Polynomial Regression is a form of linear regression in which the relationship between the independent variable x and dependent variable y is modeled as an nth degree polynomial. Polynomial regression fits a nonlinear relationship between the value of x and the corresponding conditional mean of y, denoted E(y |x), and has been used to describe nonlinear phenomena such as the growth rate of tissues, the distribution of carbon isotopes in lake sediments, and the progression of disease epidemics.


<img src="https://upload.wikimedia.org/wikiversity/en/3/35/PolynomialFunctionsGraph.PNG" alt="Polynomial Functions Graph" width="600" >


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import operator

# Step 1: Create Dummy Data
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + np.random.normal(-3, 3, 20)

# Reshaping for the model
# Reshape x and y to be column vectors (n_samples, 1) for compatibility with scikit-learn models
x = x[:, np.newaxis]
y = y[:, np.newaxis]

# plot x and y
sns.scatterplot(x=x.flatten(), y=y.flatten(), color='blue')
```

## Polynomial Regression with 1 degree

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

degree = 1
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

# evaluate model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
print("Mean Squared Error:", mean_squared_error(y, y_poly_pred))
print("R^2 Score:", r2_score(y, y_poly_pred))
print("Mean Absolute Error:", mean_absolute_error(y, y_poly_pred))
print("Mean Absolute Percentage Error:", mean_absolute_percentage_error(y, y_poly_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_poly_pred)))

# plot the polynomial regression line
plt.scatter(x, y, color='blue')
plt.plot(x, y_poly_pred, color='magenta')
plt.title('Polynomial Regression with 1 Degree')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Polynomial Regression with 2 degree

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

degree = 2
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

# evaluate model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
print("Mean Squared Error:", mean_squared_error(y, y_poly_pred))
print("R^2 Score:", r2_score(y, y_poly_pred))
print("Mean Absolute Error:", mean_absolute_error(y, y_poly_pred))
print("Mean Absolute Percentage Error:", mean_absolute_percentage_error(y, y_poly_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_poly_pred)))

# plot the polynomial regression line
plt.scatter(x, y, color='blue')
sorted_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x, y_poly_pred), key=sorted_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='magenta')
plt.title('Polynomial Regression with 2 Degree')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
