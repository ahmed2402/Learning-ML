```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
#load the data
df = sns.load_dataset('tips')
df.head()
```

```python
#create a regression line plot
sns.lmplot(y='tip', x='total_bill', data=df)
```

```python
#seprate the X and y features
X = df[['total_bill']]
y= df[['tip']]
```

```python
#load the model and fit values
model =LinearRegression()
model.fit(X,y)
```

```python
model.predict([[50]])
```
